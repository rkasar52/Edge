```python
import pandas as pd
```

```python
cluster_data = pd.read_csv("station_cluster.csv")
```

```python
cluster_data.shape
```

```python
# Don't include the last column
cluster_subset = cluster_data.iloc[:, 0:13]
cluster_subset.head()
```

```python
cluster_unpivot = pd.melt(cluster_subset, id_vars=['Number', 'Name', "Latitude", "Longitude", "District", "Public", "Total docks"], var_name='Cluster', value_name='value')
```

```python
cluster_unpivot.head()
```

```python
# Drop rows where the value of the cluster is 0, those are not useful
cluster_unpivot = cluster_unpivot[cluster_unpivot.value != 0]
# Drop the value column because we only care about the cluster number
cluster_unpivot = cluster_unpivot.drop(['value'], axis=1)
```
